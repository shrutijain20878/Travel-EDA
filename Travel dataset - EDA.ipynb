```python
# data is either numerical or categorical
# numerical can be discrete or continuous
# discrete data though its numerical but should be treated as categorical to do the analysis
# for model building all the categorical data should be converted to numbers
```

```python
cat_features = x.select_dtypes(include='object').columns
num_features = x.select_dtypes(exclude='object').columns
```

```python
num_features
```

```python
# scaling -- numerical feature
# categorical -- encoding
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler # both are transforming the data
from sklearn.compose import ColumnTransformer 
# column transformer applies transformation to columns of an array or pandas dataframe
```

```python
numeric_transformer = StandardScaler()
oh_transformer = OneHotEncoder(drop='first')
preprocessor = ColumnTransformer([
    ('OneHotEncoder',oh_transformer, cat_features),
    ('StandardScaler', numeric_transformer, num_features)
])
```

```python
preprocessor
```

```python
X_train = preprocessor.fit_transform(X_train)
pd.DataFrame(X_train)
```
